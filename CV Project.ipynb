```python
import cv2
import matplotlib.pyplot as plt
import os

def load_and_preprocess(image_path):
    img = cv2.imread(image_path)
    gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    blurred = cv2.GaussianBlur(gray, (5,5), 0)
    return img, gray, blurred

image_filenames = ["C:\\Users\\manasa\\Desktop\\img1.jpg", "C:\\Users\\manasa\\Desktop\\image3.gif", "C:\\Users\\manasa\\Desktop\\image2.gif"]

os.makedirs('results', exist_ok=True)

for image_filename in image_filenames:
    img, gray, blurred = load_and_preprocess(image_filename)
    
    cv2.imwrite(f'results/original_{image_filename}', img)
    cv2.imwrite(f'results/gray_{image_filename}', gray)
    cv2.imwrite(f'results/blurred_{image_filename}', blurred)
    
    plt.figure(figsize=(15,5))
    plt.suptitle(image_filename, fontsize=16)
    plt.subplot(1,3,1), plt.title('Original'), plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB)), plt.axis('off')
    plt.subplot(1,3,2), plt.title('Grayscale'), plt.imshow(gray, cmap='gray'), plt.axis('off')
    plt.subplot(1,3,3), plt.title('Blurred'), plt.imshow(blurred, cmap='gray'), plt.axis('off')
    plt.show()
```
